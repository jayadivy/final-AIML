```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X,_ = make_blobs(n_samples=500, centers=3, cluster_std=0.6,
random_state=0)
```

```python
db = DBSCAN(eps=0.5, min_samples=5)
y_db = db.fit_predict(X)
```

```python
n_clusters_ = len(set(y_db))-(1 if -1 in y_db else 0)
```

```python
# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[y_db == 0][:, 0], X[y_db == 0][:, 1], c='blue', marker='o',label='cluster1')
plt.scatter(X[y_db == 1][:, 0], X[y_db == 1][:, 1], c='red', marker='s',label='Cluster2')
plt.scatter(X[y_db == 2][:, 0], X[y_db == 2][:, 1], c='green', marker='^',label='Cluster3')
plt.scatter(X[y_db == -1][:, 0], X[y_db == -1][:, 1], c='gray', marker='X',label='Noise')
plt.legend(loc='best')
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
